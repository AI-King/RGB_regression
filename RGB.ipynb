```python

# import the modules
import os
from os import listdir
data_file = []
data_img = []
ds  = []
# get the path/directory
folder_dir = r'C:\Users\z_ref_analyst1\Desktop\CCR_Img_regression\test'
for images in os.listdir(folder_dir):
       # if (images.endswith(".jpg")):
          im = Image.open(images)
          stat = ImageStat.Stat(im)
          data_file.append(stat.mean)
          data_img.append(images)
          data = print(stat.mean, images)
```

```python
data_file_fr = pd.DataFrame(data=data_file, columns=['R', 'G', 'B'])
data_file_fr.head()
```

```python
data_img_fr = pd.DataFrame(data = data_img, columns = ['files'])
data_img_fr.head()
```

```python
output_frame = pd.concat([data_file_fr, data_img_fr], axis = 1)
output_frame.head()
```
